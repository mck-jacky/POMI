```python
import random
import os
from random import choice
import statistics
import sys
import pandas as pd
```

```python
event_types = ['Music', 'Performing & Visual Art', 'Seasonal', 'Health', 'Hobbies', 'Business', 'Food & Drink', 'Sports & Fitness']
```

```python
def calculate_actual_attendance(num_tickets_left, num_tickets_available, precipitation, ticket_price):
    tickets_sold = num_tickets_available - num_tickets_left
    if precipitation > 50:  # Bad weather
        return random.randint(int(0.3 * tickets_sold), int(0.7 * tickets_sold))
    else:  # Good weather
        if ticket_price < 30:  # Low-priced tickets
            return random.randint(int(0.3 * tickets_sold), int(0.6 * tickets_sold))
        else:
            return random.randint(int(0.7 * tickets_sold), tickets_sold)
```

```python
available_event_ids = set(range(31, 1001))
data = []
used_event_ids = set()

for _ in range(500):
    event_id = random.choice(list(available_event_ids))
    used_event_ids.add(event_id)
    available_event_ids.remove(event_id)

    creator_id = f'user{random.randint(1, 10)}@example.com'
    event_type = random.choice(event_types)
    ticket_price = random.randint(10, 50)
    num_tickets_available = random.randint(50, 200)
    num_tickets_left = random.randint(0, num_tickets_available)
    host = f'Host{random.randint(1, 5)}'
    is_online = random.choice([True, False])

    tmp_min, tmp_max = random.uniform(0, 20), random.uniform(20, 40)
    tmp_mid = (tmp_min + tmp_max) / 2
    precipitation = random.uniform(0, 50)
    wind_speed = random.uniform(0, 20)

    if not is_online:
        if precipitation > 50:
            num_tickets_left = num_tickets_available
        elif precipitation > 25:
            num_tickets_left = num_tickets_available - random.randint(30, 50)
        elif precipitation > 10:
            num_tickets_left = num_tickets_available - random.randint(10, 30)
        else:
            num_tickets_left = num_tickets_available - random.randint(0, 10)

    actual_attendance = calculate_actual_attendance(num_tickets_left, num_tickets_available, precipitation, ticket_price)

    past_event_stats = {
        'creator_id': creator_id,
        'event_id': event_id,
        'event_type': event_type,
        'ticket_price': ticket_price,
        'num_tickets_available': num_tickets_available,
        'num_tickets_sold': int(num_tickets_available) - int(num_tickets_left),
        'host': host,
        'is_online': is_online,
        'tmp_mid': tmp_mid,
        'tmp_min': tmp_min,
        'tmp_max': tmp_max,
        'precipitation': precipitation,
        'wind_speed': wind_speed,
        'actual_attendance': actual_attendance
    }

    data.append(past_event_stats)

df = pd.DataFrame(data)
```
